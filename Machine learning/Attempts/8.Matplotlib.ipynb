```python
#Visiualizing errors

x=np.linspace(0,10,50)
dy=0.08
dx=0.1
y=np.sin(x)

plt.errorbar(x,y,yerr=dy,xerr=dx, fmt='.k')
```

```python
x=np.linspace(0,10,50)
dy=0.8
y=np.sin(x)+dy*np.random.randn(50)

plt.errorbar(x,y,yerr=dy,fmt='.k')
```

```python
plt.errorbar(x,y,yerr=dy,fmt='o', color='black', ecolor='lightgray' )
```

```python
plt.errorbar(x,y,yerr=dy,fmt='ok', ecolor='lightgray' )
```

```python
plt.errorbar(x,y,yerr=dy,fmt='o', color='black', ecolor='lightgray',elinewidth=3)
```
