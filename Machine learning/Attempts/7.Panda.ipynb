```python
#Group By 를 이용한 계산및 요약통계

german_sample=german[['Credit Amount','Type of apartment']]
german_sample
```

```python
german_grouped=german_sample['Credit Amount'].groupby(german_sample['Type of apartment'])
```

```python
german_grouped.mean()
```

```python
german_grouped.max()
```

```python
german_sample
```

```python
german_sample=german[['Type of apartment','Sex & Marital Status','Credit Amount']]
```

```python
german_sample.groupby('Type of apartment')
```

```python
for type, group in german_sample.groupby('Type of apartment'):#'Type of apartment
#applied to type and group
  print(type)
  print(group)
```

```python
for type, group in german_sample.groupby('Type of apartment'):#'Type of apartment
#applied to type and group
  print(type)
  print(group.head())#empty head(), show the first five
```

```python
for (type, sex), group in german_sample.groupby(['Type of apartment','Sex & Marital Status']):
  print(type, sex)
  print(group.head(n=3)) #순서 중요 (type, group 순서 중요)
```

```python
births=pd.read_csv('https://raw.githubusercontent.com/jakevdp/data-CDCbirths/master/births.csv')
```

```python
births.shape
```

```python
births.head
```

```python
births['year']
```

```python
births['year']/10
```

```python
births['year']//10 #>> '//' means you divid an get rid of the remainder
```

```python
births['year']//10*10 #>> now it is the decades
```

```python
births['decade']=births['year']//10*10
```

```python
births.head() #now 'decade' column has been added
```

```python
births.pivot_table('births',index='decade',columns='gender',aggfunc='sum')
```

```python
import matplotlib.pyplot as plt
```

```python
births.pivot_table('births',index='decade',columns='gender',aggfunc='sum').plot()
```
