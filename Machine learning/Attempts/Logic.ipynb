```python
import numpy as np
import matplotlib.pyplot as plt

#making the AND gate
def AND(a,b):
  input=np.array([a,b])

  #weighted value
  weights=np.array([0.4,0.4])
  bias=-0.6

  #printed value
  value=np.sum(input*weights)+bias

  #return value

  if value<=0:
    return 0
  else:
    return 1
```

```python
print(AND(0,0))
print(AND(0,1))
print(AND(1,0))
print(AND(1,1))
```

```python
#making the OR gate
def OR(a,b):
  input=np.array([a,b])

  #weighted value
  weights=np.array([0.4,0.4])
  bias=-0.3

  #printed value
  value=np.sum(input*weights)+bias

  #return value

  if value<=0:
    return 0
  else:
    return 1


print(OR(0,0))
print(OR(0,1))
print(OR(1,0))
print(OR(1,1))
```

```python
#making NAnd Statement
def NAND(a,b):
  input=np.array([a,b])

  #weighted value
  weights=np.array([-0.6,-0.6])
  bias=0.7

  #printed value
  value=np.sum(input*weights)+bias

  #return value

  if value<=0:
    return 0
  else:
    return 1


print(NAND(0,0))
print(NAND(0,1))
print(NAND(1,0))
print(NAND(1,1))
```

```python
def XOR(x1, x2):
  s1 = NAND(x1, x2)
  s2 = OR(x1, x2)
  y = AND(s1, s2)

  return y
```

```python
print(XOR(0,0))
print(XOR(0,1))
print(XOR(1,0))
print(XOR(1,1))
```

```python
#Activation Function


#Step Function

def step_function(x):
  if x>0:
    return 1
  else:
    return 0
```

```python
def step_function_for_numpy(x):
  y=x>0
  value=y.astype(np.int)

  return value
```

```python
print(step_function(-3))
print(step_function(5))
```

```python
#Sigmoid Function

def sigmoid(x):
  value=1/(1+np.exp(-x))
  return value


print(sigmoid(3))
print(sigmoid(-3))
```

```python
#Softmax(a)

def Softmax(a):
  exp_a=np.exp(a)
  sum_exp_a=np.sum(exp_a)
  y=exp_a/sum_exp_a

  return y
```

```python
a=np.array([0.3,0.2,3.0,-1.2])
print(Softmax(a))
print(np.sum(Softmax(a)))
```
