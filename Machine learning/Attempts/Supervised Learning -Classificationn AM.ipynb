```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import sklearn
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.datasets import load_iris
iris=load_iris()
```

```python
print(iris.DESCR)
```

```python
data=iris.data
label=iris.target
columns=iris.feature_names
```

```python
data=pd.DataFrame(data,columns=columns)
data.head()
```

```python
data.shape
```

```python
#data preprocessing

from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(
    data,
    label,
    test_size=0.2,
    random_state=2022
)


#logistic regression
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
```

```python
lr.fit(X_train,y_train)
y_pred=lr.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print('logisitic regression accuracy:{:.2f}'.format(accuracy_score(y_test,y_pred)))
```

```python
#support vector 
from sklearn.svm import SVC
svc=SVC()
```

```python
svc.fit(X_train,y_train)
y_pred=svc.predict(X_test)
print('SVC accuracy:{:.2f}'.format(accuracy_score(y_test,y_pred)))
```

```python
#Decision Tree
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier(max_depth=5)

dt.fit(X_train,y_train)
y_pred=dt.predict(X_test)
print('Decision Tree accuracy:{:.2f}'.format(accuracy_score(y_test,y_pred)))
```

```python
#Random Forest
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier(max_depth=5)

rf.fit(X_train,y_train)
y_pred=rf.predict(X_test)

print('Random Forest Accuracy {:.2f}'.format(accuracy_score(y_test,y_pred)))
```
