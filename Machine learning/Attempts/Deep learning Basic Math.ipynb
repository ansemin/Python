```python
import math 
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
```

```python
#linear equation
def linear_function(x):
  a=0.5
  b=2

  return a*x+b
```

```python
print(linear_function(5))
```

```python
np.arange(-5,5,0.1) #from -5 to 5, in 0.1 steps
```

```python
x=np.arange(-5,5,0.1)
y=linear_function(x)
```

```python
y
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Function')
```

```python
#Quadratic function
def quadratic_function(x):
  a=1
  b=-1
  c=-2

  return a*x**2+b*x+c
```

```python
print(quadratic_function(2))
```

```python
y=quadratic_function(x)

plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Quadratic function')
```

```python
#Cubic func

def cubic_function(x):
  a=1
  b=2
  c=3
  d=4

  return a*x**3+b*x**2+c*x+d

y=cubic_function(x)

plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('cubic function')
```

```python
#Finding the minimum 

def my_func(x):
  a=1
  b=-3
  c=10
  return a*x**2+b*x+c

x=np.arange(-10,10,0.1)
y=my_func(x)

plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('My function') 
```

```python
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('My function') 
plt.scatter(1.5,my_func(1.5))
```

```python
#Logorithmic function 

def exponential_function(x):
   a=4
   return a**x

print(exponential_function(4))
print(exponential_function(0))
```

```python
x=np.arange(-3,2,0.1)
y=exponential_function(x)

plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Log Function') 
```
